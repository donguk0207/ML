```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:4].values,
                 df["target"],
                 test_size=0.3,
                 random_state=27)
```

## 학습알고리즘 선택(모델선택)
Y(Class, Target) value ==> categorical data ==> classification model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
print(knn)
```

```python
knn.fit(x_train, y_train)
```

```python
#예측
y_pred = knn.predict(x_test)
y_pred
```

```python
#모델의 정확도
acc_result = knn.score(x_test, y_test)
acc_result
```

```python
#1. load data
#2. preprocessing
#3. train_test_split ==> hold-out dataset
#4. baseline model(A) fit ==> model tunning(Hyper param, ) : A'
#5. target model(B) fit ==> model tunning(Hyper param, ) : B'
#6. performance test : using metric(MSE, MAE(*), accuracy, precision, recall, auc_score(*))

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(x_test, y_test)
score_8 = knn.score(x_test, y_test)
print(score_8)
```

```python
knn_acc = list()

for k in range(2, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    acc_result = knn.score(x_test, y_test)
    item = {
        "K": k,
        "acc": acc_result
    }
    knn_acc.append(item)
```

```python
for item in knn_acc:
    print(item)
```

```python
import matplotlib.pyplot as plt
x, y = list(), list()

for item in knn_acc:
    x.append(item['K'])
    y.append(item['acc'])

plt.plot(x, y)
plt.show()
```

```python
knn_acc = list()

acc_train = []
acc_test = []
k_list = range(1, 100, 2)


for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    acc_train.append(knn.score(x_train, y_train))
    acc_test.append(knn.score(x_test, y_test))

```

```python
plt.plot(k_list, acc_train, 'b--')
plt.plot(k_list, acc_test, 'g')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(k_list, acc_train, 'b--', label="acc_train")
plt.plot(k_list, acc_test, 'g', label="acc_test")
plt.xlabel("k")
plt.ylabel("acc")
plt.ylim(0.8, 1.0)
plt.legend()
plt.show()
```
