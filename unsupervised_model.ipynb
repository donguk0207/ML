# Unsupervised Learning 

## 계층적 군집분석(Hierarchical Clustering)

- data scaling
- linkage()로 distance matrix 산출
- dendrogram을 통해서 어떤 것들이 유사한지 확인
- 군집결정: fcluster()
- 시각화

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris["data"], columns=iris["feature_names"])
df.head()
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
scaled_data = std_scaler.fit_transform(df)
scaled_data
```

```python
# Distance Matrix
from scipy.cluster.hierarchy import linkage, fcluster
```

```python
#sparse matrix
distance_matrix = linkage(scaled_data, method="ward", metric="euclidean")
distance_matrix 
```

- 덴드로그램 그리기 

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize=(30,10))
dn = dendrogram(distance_matrix, labels=iris["target"], leaf_rotation=90, leaf_font_size=20)
plt.show()
```

```python
df.head()
```

```python
# criterion : ['maxclust','distance'] 
# maxclust = 만들 수 있는 만큼 최대한 만들기 - 이 경우 앞의 값은 cuttree의 수, 군집의 최종 수
# distance = 어느 정도의 거리 이상이면 군집으로 만들지 말라는 의미 (13이상) - 이 경우 앞의 값은 distance threshold. 
df["cluster_label"] = fcluster(distance_matrix, 3, criterion="maxclust")
df.head()
```

```python
df["cluster_label"].value_counts()
```

```python
# 확인을 위해서 시각화
df["label"] = iris["target"]
df.head()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(16,8))
# c option: color sequence
axes[0].scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=df["label"])
axes[1].scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=df["cluster_label"])

axes[0].set_title("Real", fontsize=20)
axes[1].set_title("Hierarchical Clustering", fontsize=20)

plt.show()
```

## 비계층적 군집분석(K-means) 

```python
from sklearn.cluster import KMeans
```

```python
df.drop(["cluster_label"], axis=1, inplace=True)
df.head()
```

```python
cols = list(df.columns[:-1])
data = df.loc[:, cols]
```

```python
model = KMeans(n_clusters=3).fit(data)
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=df["label"])

center_x = model.cluster_centers_[:, 0] # sepal length centroid value
center_y = model.cluster_centers_[:, 1] # sepal width centroid value
plt.scatter(center_x, center_y, c="red", marker="D") # marker="D" -> diamond
plt.show()
```

```python
# Optimal K -> distortion -> Inertia Value(응집도 - 거리의 합)
# 거리의 합 => 값이 작을 수록 응집도가 높다

inertias = []

for i in range(1,11):
    k_means = KMeans(n_clusters=i, random_state=47)
    k_means.fit(data)
    inertias.append(k_means.inertia_)

plt.plot(range(1,11), inertias, marker="o")
plt.xlabel("# of clusters")
plt.ylabel("Distortion")
plt.show()
```

## 주성분분석(PCA)

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
# PCA 
# Scaling
# 주성분분석

from sklearn.preprocessing import StandardScaler

X_train = df.iloc[:, :4]
z_scaled = StandardScaler().fit_transform(X_train)
z_scaled
```

```python
from sklearn.decomposition import PCA

# n_components는 주성분을 의미. 주성분 2개를 만들어달라는 의미. 
pca = PCA(n_components=2)
pca.fit(z_scaled)
```

```python
pca.components_ #주성분 Principle Compnent 1, Principle Compnent2 --> 고유벡터 
```

```python
#위의 고유벡터(pc1, pc2)에 데이터를 정사영 시키자.
X_pca = pca.transform(z_scaled)
X_pca.shape
```

```python
#검증 : 각 주성분이 분산을 얼마나 잘 나타냈는지를 정량화
pca.explained_variance_ratio_
```

```python
# 4개의 feature --> 2개의 축으로 정사영(projection) 시켰을 때, 설명되는 분산
np.sum(pca.explained_variance_ratio_)
```

```python
# projection된 데이터를 데이터프레임으로 변경한 후, 모델에 적용
pca_cols = ["pca_com_1", "pca_com_2"]
pca_df = pd.DataFrame(data=X_pca, columns=pca_cols)
pca_df["target"] = df["label"]
pca_df.head()
```

```python
# model test
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df.columns
```

```python
X = df.iloc[:, :-1]
y = df["label"]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=27)
```

```python
dt = DecisionTreeClassifier(random_state=27).fit(x_train, y_train)
```

```python
dt
```

```python
y_pred = dt.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#비교를 위해서 pca data를 넣어보자

X = pca_df.iloc[:, :-1]
y = pca_df["target"]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=27)
dt = DecisionTreeClassifier(random_state=27).fit(x_train, y_train)
y_pred = dt.predict(x_test)
print(classification_report(y_test, y_pred))
```

## Association Rule 연관분석

```python
import mlxtend #aprior, Fp-Growth 알고리즘을 사용하기 위해서
```

```python
data = np.array([
  ['식빵', '버터', '우유'],
  ['바게트빵', '소금', '올리브유'],
  ['식빵', '버터', '우유', '잼'],
  ['치즈케익', '바게트빵', '식빵'],
  ['식빵', '잼', '샐러드유'],
  ['치즈케익', '잼', '치즈'],
  ['치즈', '우유'],
  ['식빵'],
  ['식빵', '버터', '우유'],
  ['바게트빵']
], dtype=object)
```

```python
data
```

```python
#Apriori 알고리즘
from mlxtend.preprocessing import TransactionEncoder
```

```python
trans_enc = TransactionEncoder()
trans_arr = trans_enc.fit_transform(data)
trans_arr

#빈발도를 구한것임, 빈발도를 구하면서 n by n 을 맞춰준거임
```

```python
df = pd.DataFrame(trans_arr, columns=trans_enc.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# apriori 알고리즘을 쓸때 설정해야하는 하이퍼파라미터는 최소지지도 : min_support

apriori(df, min_support=0.3, use_colnames=True)
```

```python
association_rules(apriori(df, min_support=0.3, use_colnames=True), num_itemsets=8)
#antecedents = 선행 (x)
#consequents = 후행 (y)
```

```python
# Fp-growth
# pip install mlxtend --upgrade
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.3, use_colnames=True)
```

```python
association_rules(fpgrowth(df, min_support=0.3, use_colnames=True), num_itemsets=8)
```
